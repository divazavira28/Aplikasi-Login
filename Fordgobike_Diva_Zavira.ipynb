```python
fordgobike_copy = fordgobike.copy()
```

```python
fordgobike_copy.head()
```

```python
fordgobike_copy = fordgobike_copy.dropna(how='any', subset=['start_station_id'])
fordgobike_copy = fordgobike_copy.dropna(how='any', subset=['start_station_name'])
fordgobike_copy = fordgobike_copy.dropna(how='any', subset=['end_station_id'])
fordgobike_copy = fordgobike_copy.dropna(how='any', subset=['end_station_name'])
fordgobike_copy = fordgobike_copy.dropna(how='any', subset=['member_birth_year'])
fordgobike_copy = fordgobike_copy.dropna(how='any', subset=['member_gender'])
```

```python
print(fordgobike_copy.isnull().sum())
print()
print(f"Jumlah Kolom : {fordgobike_copy.shape[1]}")
print(f"Jumlah Baris : {fordgobike_copy.shape[0]}")
```

```python
sns.heatmap(fordgobike_copy.isnull())
```

```python
fordgobike_copy = fordgobike_copy.drop_duplicates()
print(f"Jumlah Kolom : {fordgobike_copy.shape[1]}")
print(f"Jumlah Baris : {fordgobike_copy.shape[0]}")
```

```python
plt.figure(figsize = (8,8))
plt.boxplot(fordgobike_copy["duration_sec"])
```

```python
Q1 = fordgobike_copy["duration_sec"].quantile(0.25)
Q3 = fordgobike_copy["duration_sec"].quantile(0.75)
IQR = Q3 - Q1

outlierQ3 = Q3 + 1.5 * IQR

fordgobike_copy = fordgobike_copy[(fordgobike_copy["duration_sec"] < outlierQ3)]
fordgobike_copy["duration_sec"].describe()
```

```python
print(f"jumlah kolom data copy : {fordgobike_copy.shape[1]}")
print(f"jumlah Baris data copy : {fordgobike_copy.shape[0]}")
```

```python
plt.figure(figsize = (8,8))
plt.boxplot(fordgobike_copy["member_birth_year"])
```

```python
Q1 = fordgobike_copy["member_birth_year"].quantile(0.25)
Q3 = fordgobike_copy["member_birth_year"].quantile(0.75)
IQR = Q3 - Q1

outlierQ1 = Q1 - 1.5 * IQR

fordgobike_copy = fordgobike_copy[(fordgobike_copy["member_birth_year"] > outlierQ1)]
fordgobike_copy["member_birth_year"].describe()
```

```python
print(f"jumlah kolom data copy : {fordgobike_copy.shape[1]}")
print(f"jumlah Baris data copy : {fordgobike_copy.shape[0]}")
```
